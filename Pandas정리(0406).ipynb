## 시리즈는 무엇인가요?
데이터프레임의 컬럼(행)은 모두 시리즈입니다.

```python
# 첫번째 시리즈 생성해보기
a = pd.Series(['john','jenny','Nate','Julia','Brian'])
# 시리즈 name 은 name 설정
a.name = 'name'
```

```python
#  두번째 시리즈 생성해보기
b = pd.Series([20.30,30,40,45])
# 시리즈 name 은 age 설정
b.name ='age'
```

```python
# job 시리즈 생성해보기
c= pd. Series(['student','developer','teacher','dentist','manager'])
# 시리즈 name 은 job으로 설정
c.name = 'job'
```

### 추가) 생성한 데이터 프레임 csv 파일로 저장하는 법
데이터프레임 (dataframe)은 2차원 자료구조입니다. 로컬럼으로 엑셀 형식과 유사합니다.  
기본적으로 csv 포맷을 지원하지만, 구분자로 컬럼이 구분되어 있는 데이터는 모두 지원합니다.  
read_csv 함수로 파일을 데이터프레임으로 호출하고 저장 할 수 있습니다.

#### df.to_csv('파일경로', header=False) 

- 저장하고자 하는 데이터의 변수 -> df  
-  말한대로 추가가 되고 header를 False 값을 주면 추가되는 값에 header가 찍히지 않는다.
