## 범례 위치지정 키워드

```python
from matplotlib import font_manager, rc
font_list = font_manager.findSystemFonts()
font_list
```

```python
# enumerate: 인덱스 번호 출력, 데이터도 하나씩 출력
font_lst = []
for i, fontName in enumerate(font_list):
      a =font_manager.FontProperties(fname = fontName)
      font_lst.append([i,a.get_name()])
```

- 1.요일별 사고 건수 count

```python
data
```

```python
data['요일'].unique()
```

```python
temp =data['요일'].value_counts()
# 요일 순서로 정렬
y = temp[['월','화','수','목','금','토','일']]
y
```

```python
# 요일별 사망 교통사고 건수
# ylim 500 ~620
# x축 이름 '요일'
# y축 이름 '사고 건수'
# title '요일별 교통사고 건수'
```

```python
x = range(7)
plt.bar(x,y)
plt.xticks(x,['월','화','수','목','금','토','일'])
plt.ylim(500,620)
plt.xlabel('요일')
plt.ylabel('사고수')
plt.title('요일별 교통사고 횟수')
plt.show()
```

- 2. 차대차 사건 중 죽거나 다친 사림이 많은 발생지 시도를 알아보고 시각화 해보자

```python
data.head()
```

```python
#1. 사고유형_대분류 차대차인 행 데이터 출력
#2. 발생지 시도 접근 후 각 지역이 몇번 등장하는지 횟수 세기
```

```python
acc =data[data['사고유형_대분류'] == '차대차']
```

```python
acc['발생지시도'].value_counts().shape
```

```python
# 사상자 수, 발생지 시도만 컬럼접근
acc2=acc [['사상자수','발생지시도']].groupby('발생지시도').sum()
```

```python
acc2
```

```python
x = acc2.index
y = acc2['사상자수']

plt.figure(figsize = (10,5)) # figsize = (10,5) 10- X넓이 , 5-Y넓이
plt.bar(x,y)
plt.ylabel('사상자수')
plt.xlabel('지역')
plt.title('차대차 교통사고의 사상자 수')
plt.show()
# 이거 커널 끊고 다시 실행하면 됩니다  감사합니다
```
