### DataFrame 병합
- concat
- merge

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
print(df1)
print(df2)
print(df3)
```

```python
# concat
pd.concat([df1,df2,df3],axis=1)
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
```

```python
df4
```

```python
pd.concat([df1,df4],axis=1)
```

```python
# join 속성
# outer : 기본값
# inner : 합치는 데이터들이 동일하게 가지고있는 기준만 출력
```

```python
pd.concat([df1,df4],join = 'inner', axis =1) 
##outer는 default임으로 안써도 무방
```

```python
# ignore_index
# ignore_index = True
# 기존에 가지고있던 index를 무시하고 새로운 인덱스 번호 부여
pd.concat([df1,df4],ignore_index=True) 
```

```python
# merge
# 값을 기준으로 데이터를 합침(공통값만 머지)
```

```python
df5 = pd.DataFrame({'key' : ['K0','K2','K3','K4'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})

df6 = pd.DataFrame({'key' : ['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
```

```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6,on='key') #inner가 디폴트
```

```python
# how
# concat의 join과 동일한 기능
# outer : 전체 값 출력
# inner : 기준이 동일한 값 출력 :디폴트
# left : 먼저 적은 데이터를 기준으로 값 출력
# right : 나중에 적은 데이터를 기준으로 값 출력  (나중값을 기준으로 데이터 채움)
```

```python
df_merge = pd.merge(df5,df6,how ='right')
df_merge
```

```python
df_merge = pd.merge(df5,df6,how ='left')
df_merge
```

```python
# 결측치 채우기
# 변수.fillna(채울값)
df_merge.fillna(0)
```
